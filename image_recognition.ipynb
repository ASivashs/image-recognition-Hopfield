```python
class AsyncHopfield:

    def __init__(self):
        self.weight = [[0 for _ in range(MODEL_SIZE ** 2)] for line in range(MODEL_SIZE ** 2)]
        self.shuffled_neurons = [index for index in range(MODEL_SIZE ** 2)]
        np.random.shuffle(self.shuffled_neurons)

    def train(self, image, pre_trained=False):
        if not pre_trained:
            first = np.array(image).flatten()
            first = first.reshape(1, first.shape[0])
            first = first.T
            second = np.array(image).flatten()
            second = second.reshape(1, second.shape[0])
            mul = first @ second
            weight_matrix = np.array(self.weight) + mul
            self.weight = weight_matrix.tolist()
            for nullifying_index in range(len(self.weight)):
                self.weight [nullifying_index][nullifying_index] = 0
            with open('pre-trained/weights.npy', 'wb') as weight_file:
                np.save(weight_file, self.weight)
        else:
            with open('pre-trained/weights.npy', 'rb') as weight_file:
                self.weight = np.load(weight_file)

    def recognize(self, corrupt_image, async_mode=False):
        counter = 0
        np_wm = np.array(self.weight)
        complete = False
        next_step = np.array([corrupt_image]).flatten().T

        while not complete:
            counter += 1
            if counter > 10000:
                print('Image is not detected.')
                break
            
            previous_iteration = next_step

            multiplied_matrix = np.matmul(np_wm, next_step)

            if not async_mode:
                next_step = multiplied_matrix

            if async_mode:
                if counter == 1:
                    next_step = np.array([corrupt_image]).T.flatten()

                cycle = counter % (MODEL_SIZE ** 2)
                current_random_neuron = self.shuffled_neurons[cycle]
                next_step[current_random_neuron] = multiplied_matrix[current_random_neuron]

            for activating_index in range(len(next_step)):
                if next_step[activating_index] >= 0:
                    next_step[activating_index] = 1
                else:
                    next_step[activating_index] = -1

            # print(next_state.T.flatten().tolist())

            #if (previous_iteration == next_state).all():
            if np.array_equal(previous_iteration, next_step):
                complete = True

        return next_step.T
```
